## Encryption Using Caesar Cipher

The **caesar_str_enc function** takes two inputs, the **text** to be encrypted and an integer **s** representing the number of places to shift each letter. The function first converts the input text to upper case and removes any whitespace. It then iterates over each character in the text and applies the Caesar Cipher encryption formula to each character. The result is a new string representing the encrypted text.

```python
def caesar_str_enc(text,s):
  #Defining a variable to save the input data and coverting to upper case, if the input has some char in lowercase and also removing whitespaces
  upper_text = text.upper().replace(" ","")                               
  result = ""
  i=0
  while i < len(upper_text):
      char = upper_text[i]
      result += chr((ord(char) + s-65) % 26 + 65)
      i +=1
  return result 
```

The **caesar_str_dec** function takes two inputs, the **text** to be decrypted and an integer **s** representing the number of places the original text was shifted. The function iterates over each character in the input text and applies the Caesar Cipher decryption formula to each character. The result is a new string representing the decrypted text.

```python
def caesar_str_dec(text,s):
  result = ""
  i=0
  while i < len(text):
      #Converting the string to char
      char = text[i]
      #Formula to Decrypt the encrypted data.                     
      result += chr((ord(char) - s+65) % 26 + 65)
      i +=1
  return result 
```

In the main part of the code, the user is prompted to enter the input text and the desired shift key. The caesar_str_enc function is called with these inputs, and the resulting encrypted text is printed to the console. The caesar_str_dec function is then called with the encrypted text and the original shift key, and the resulting decrypted text is printed to the console.

```python
if __name__=="__main__":
  print('Enter your data which requires encryption: ')
  #Defining a variable to save the input for encryption by the user   
  input_str = input()                                                    
  print('\nEnter your desired key from 0 to 25: ')
  #Defining a variable to save the key input for caesar cipher by the user
  input_shift = int(input())                                             
  #Defining a variable to save the output of the called caesar encryption function
  mytest_enc = caesar_str_enc(input_str,input_shift)                     
  print("\nEncrypted data using Caesar Cipher is: ",mytest_enc)
  #Defining a variable to save the output of the called caesar decryption function.
  mytest_dec = caesar_str_dec(mytest_enc,input_shift)                    
  print("\nDecrypted Data: ",mytest_dec)
```

#Overall, this code provides a simple implementation of the Caesar Cipher and demonstrates how encryption and decryption can be achieved using simple mathematical operations. However, it is important to note that the Caesar Cipher is a relatively weak cipher and can be easily broken using modern cryptanalysis techniques. Therefore, it should not be used for secure communications in real-world applications.
